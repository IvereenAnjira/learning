### Use Case 
Are there certian countries that grow faster than others 

N.B Doing it in terms of continents because there are to many countiers to put in a visualization 

```python
df2= df.groupby('Continent')[df.columns[]].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
df2
```

```python
df3= df2.transpose()
df3
```

```python
df.columns
```

```python
df3.plot()
```

The formed table above will be used to plot a graph showing the changes in population in each continent, to achive that either of the following has to be done 

#### Reversing the range 
df2 = df.groupby('Continent')[df.columns[5:13][::-1]].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
df2

 or start counting columns backwards,
    df2 = df.groupby("Continent")[df.columns[-5:-13:-1]].mean().sort_values(by='2022 Population', ascending=False)

df3.plot().invert_xaxis()
